```python
import pandas as pd
df=pd.read_csv("bike.csv")
df
```

```python
# a brief summary for the datbase
df.describe().T
```

```python
#check for the null values
df.isna().apply(pd.value_counts)
```

```python
#Info about the data shape of data, type of individual columns
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
plt.boxplot(df["duration_sec"])
plt.figure(figsize= (20,20))
plt.subplot(3,3,1)
plt.hist(df["duration_sec"], color='lightblue', edgecolor = 'black')
plt.show()
plt.show()
```

```python
isnull_number = []
for i in df.columns:
    x = df[i].isnull().sum()
    isnull_number.append(x)

pd.DataFrame(isnull_number, index = df.columns, columns = ['Total Missing Values'])
```

```python
unique_number = []
for i in df.columns:
    x = df[i].nunique()
    unique_number.append(x)

pd.DataFrame(unique_number, index = df.columns, columns = ['Total Unique Values'])
```

```python
# Plots to see the distribution of the continuous features individually

plt.figure(figsize= (20,15))
plt.subplot(3,3,1)
plt.hist(df.start_station_id, color='lightblue', edgecolor = 'black', alpha = 0.7)
plt.xlabel('start_station_id')

plt.subplot(3,3,2)
plt.hist(df.end_station_id, color='lightblue', edgecolor = 'black', alpha = 0.7)
plt.xlabel('end_station_id')


plt.show()
#to show the diff between start and station ids
```
